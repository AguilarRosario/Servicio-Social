```python
datapath = 'C:\\Users\\HP desktop\\Documents\\Trimestre 19I\\Servicio social\\images\\' #Ruta donde se desea guardar las gráficas
```

```python
plt.figure(figsize=(9,7))
plt.subplot(2,1,1)
plt.semilogx(volt[:,0],20*np.log10(volt[:,1]/3.5),linewidth=3)
# plt.title("Magnitud de respuesta en frecuencia\nn=4")
plt.ylabel('Magnitud [dB]',fontsize=15)
fc=np.where(volt==2.56) # Ubicar posición donde se encuentra la fc en la matriz volt
f = fc[0] #fila
c = fc[1] #Columna
punto = plt.plot(volt[f,0], 20*np.log10(volt[f,c]/3.5), 'ro') # Muestra la frecuencia de corte en la gráfica de magnitud
plt.grid(True)

plt.subplot(2,1,2)
plt.semilogx(volt[:,0],volt[:,2],linewidth=3)
plt.ylabel('Fase [grados]',fontsize=15)
plt.xlabel('Frecuencia [Hz]',fontsize=15)
punto = plt.plot(volt[f,0], volt[f,2], 'ro') # Muestra la frecuencia de corte en la gráfica de fase
plt.grid(True)
plt.savefig(datapath+'RF_PA4')
```
