```python
import numpy             as np
import matplotlib.pyplot as plt
```

```python
"""PASA BAJAS"""
"""Señal de entrada: senoidal 1.88V"""
#[Hz,Volts,Fase]
volt = np.array([[0.6,5.2,0],[1.9,5.2,0],[4.1,5.12,0],[6.0,5.12,-11],[8.,5.12,-12],[10,5.08,-15],[20,4.88,-27],[30,4.64,-40],[40,4.40,-53],[50,4.08,-66],[55,3.88,-73],[56,3.88,-73.5],[57,3.80,-75],[58,3.80,-75.7],[59,3.80,-76.9],[60,3.72,-80],[61,3.64,-80],[62,3.60,-83],[63,3.56,-82.4],[64,3.48,-83.2],[65,3.48,-83],[70,3.28,-91],[80,2.20,-103],[90,2.40,-109],[100,2.12,-116],[200,.640,-142],[300,0.320,-164],[400,0.200,-170]])
```

```python
datapath = 'C:\\Users\\HP desktop\\Documents\\Trimestre 19I\\Servicio social\\images\\' #Ruta donde se desea guardar las gráficas
```

```python
plt.figure(figsize=(9,7))
plt.subplot(2,1,1)
plt.semilogx(volt[:,0],20*np.log10(volt[:,1]/5.2),'g',linewidth = 3)
# plt.title("Respuesta en frecuencia\nn=2")
plt.ylabel('Magnitud [dB]',fontsize = 15)
# Punto a señalar en la primera gráfica
fc=np.where(volt==3.64) # Ubicar posición donde se encuentra la fc en la matriz volt
f = fc[0] #fila
c = fc[1] #Columna
punto = plt.plot(volt[f,0], 20*np.log10(volt[f,c]/5.2), 'ro') # Muestra la frecuencia de corte en la gráfica de magnitud
plt.grid(True)

plt.subplot(2,1,2)
plt.semilogx(volt[:,0],volt[:,2], linewidth = 3)
plt.ylabel('Fase [grados]',fontsize = 15)
plt.xlabel('Frecuencia [Hz]',fontsize = 15)
punto = plt.plot(volt[f,0], volt[f,2], 'ro') # Muestra la frecuencia de corte en la gráfica de fase
plt.grid(True)
plt.savefig(datapath+'RF_PB2')
```
